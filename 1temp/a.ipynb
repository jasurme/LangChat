```python
def read_lines_in_range(file_path, start_line, end_line):
    """
    Reads lines from a text file within a specified range (inclusive).
    Line numbers are 1-based.
    """
    selected_lines = []
    # Open the file and iterate through its lines
    with open(file_path, 'r') as file:
        # Enumerate adds a counter to the iterator
        for line_num, line in enumerate(file, 1):
            if start_line <= line_num <= end_line:
                selected_lines.append(line.strip('\n'))
            elif line_num > end_line:
                # Stop reading once the end range is exceeded
                break
    return selected_lines

# Example usage:
file_path = '../files/llms.txt' # Replace with your file name/path
start_line = 3
end_line = 6

try:
    lines = read_lines_in_range(file_path, start_line, end_line)
    for line in lines:
        print(line)
except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found.")

```
